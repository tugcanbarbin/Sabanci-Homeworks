```python
import numpy as np
import matplotlib.pyplot as plt


sample_size = 100

X = np.linspace(0,10,sample_size)
slope = 5
bias = 20

y = X * slope + np.random.randn(sample_size) * 10 + bias
print(y)
print(X)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.xlabel('X labels')
plt.ylabel('y labels')
plt.title('dataset')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

first_linear_Regression = LinearRegression()
first_linear_Regression.fit(X.reshape(-1,1),y)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
y_pred = first_linear_Regression.predict(X.reshape(-1,1))
print('Coefficients: \n', first_linear_Regression.coef_)
print('Bias', first_linear_Regression.intercept_)
print('mean absolute error : %.2f' % mean_absolute_error(y,y_pred))
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,y_pred, 'red')
```

```python
#polynomial model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X.reshape(-1,1))
```

```python
print(X.reshape(-1,1).shape)
X_poly.shape, X.shape
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y.reshape(-1,1))
y_pred = lin_reg_2.predict(X_poly)
print('Coefficients: \n', lin_reg_2.coef_)
print('Bias', lin_reg_2.intercept_)
print('mean absolute error : %.2f' % mean_absolute_error(y,y_pred))
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,y_pred, 'red')
```

```python
#degreelere gore error degisimi 
poly_reg = PolynomialFeatures(degree = 15)
X_poly = poly_reg.fit_transform(X.reshape(-1,1))
print(X.reshape(-1,1).shape)
X_poly.shape, X.shape
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y.reshape(-1,1))
y_pred = lin_reg_2.predict(X_poly)
print('Coefficients: \n', lin_reg_2.coef_)
print('Bias', lin_reg_2.intercept_)
print('mean absolute error : %.2f' % mean_absolute_error(y,y_pred))
plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,y_pred, 'red')
```
